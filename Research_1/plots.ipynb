```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load and plot loss_single.npy, loss_trad_batched.npy and loss_tt_batched.npy
loss_single = np.load('data/loss_single.npy')
loss_trad_batched = np.load('data/loss_trad_batched.npy')
loss_tt_batched = np.load('data/loss_tt_batched.npy')

plt.plot(loss_single, label='Single', alpha=0.8)
plt.plot(loss_trad_batched, label='Trad batched 32', alpha=0.8)
plt.plot(loss_tt_batched, label='TT Batched 32', alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# load and plot accuracy_single.npy, accuracy_trad_batched.npy and accuracy_tt_batched.npy
accuracy_single = np.load('data/accuracy_single.npy')
accuracy_trad_batched = np.load('data/accuracy_trad_batched.npy')
accuracy_tt_batched = np.load('data/accuracy_tt_batched.npy')

plt.plot(accuracy_single, label='Single', alpha=0.8)
plt.plot(accuracy_trad_batched, label='Trad Batched 32', alpha=0.8)
plt.plot(accuracy_tt_batched, label='TT Batched 32', alpha=0.8)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
